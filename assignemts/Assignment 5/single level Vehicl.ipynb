```python
#parent class 
class Vehicl:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model

    def show(self):
        print("Brand:",self.brand)
        print("model:",self.model)


#child class
class Car(Vehicl):
    def __init__(self,brand,model,seats):
        super().__init__(brand,model)
        self.seats=seats

    def show(self):
        super().show()
        print("seats:",self.seats)
```

```python
C=Car("Honda" ,"Civic",4)
C.show()
```
