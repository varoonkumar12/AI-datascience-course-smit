```python
class Animal:
    def make_sound(self):
        print("animal sound")

#child class 
class Dog(Animal):
    def make_sound(self):
        print("bark")
```

```python
A=Animal()
A.make_sound()
```

```python
D=Dog()
D.make_sound()
```

```python
#overriding
Animal_ref=Dog()
Animal_ref.make_sound()
```
