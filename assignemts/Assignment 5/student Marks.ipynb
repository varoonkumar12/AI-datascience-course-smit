```python
class Student_Marks:
    def __init__(self,name,roll_no,marks=0):
        self.__name=name
        self.__roll_no=roll_no
        self.__marks= 0
        
        self.set_marks(marks)

   #setter method
    def set_marks(self,marks):
        if 0<= marks<=100:
            self.__marks=marks
        else:
            print("Invalid marks!")
            self.__marks=0

  #get method
    def get_name(self):
        return self.__name

    def get_roll_no(self):
        return self.__roll_no

    def get_marks(self):
         return self.__marks
                  
   
    
```

```python
s1=Student_Marks("varoon",44,100)
print("Name:",s1.get_name())
print("roll_no:",s1.get_roll_no())
print("marks:",s1.get_marks())

s1.set_marks(111)
print("marks are invalid",s1.get_marks())
      
```
