```python
#Grand parents
class GrandParents:
    def family_name(self):
        print("Family name is Akhani" )

#Parents
class Parents(GrandParents):
    def Family_Occupution(self):
        print("Family Occupution is Business")

#child class
class Child(Parents):
    def Hobby(self):
        print(" Hobby is circket")
```

```python
c = Child()
c.family_name()
c.Family_Occupution()
c.Hobby()

```
