```python
class ShoppingCart:
    def __init__(self):
        self.__item= []
    def add_item(self,item):
        if item not in self.__item:
            self.__item.append(item)
        else:
            print("item already in cart")

    def remove_item(self,item):
        if item in self.__item:
            self.__item.remove(item)
        else:
            print("item not found")

    def view_cart(self):
        return self.__item
```

```python
cart=ShoppingCart()
cart.add_item("Apple")
cart.add_item("cheku")
cart.add_item("papaya")
print("Cart:",cart.view_cart())

```

```python
cart.remove_item("papaya")
print("cart:",cart.view_cart())
```
