```python
class Password_Manager:
    def __init__(self,username,password):
        self.username=username
        self.__password=password

    #set password
    def set_password(self,old,new):
        if self.__password==old:
            self.__password=new
            print("password update succesfully")
        else:
            print("old password dont match!")

    #user name method
    def __check_username(self,name):
        return self.username==name
    #user password method
    def __check_password(self,passd):
        return self.__password==passd

   # login verify method
    def verify_method(self,name,passd):
        if self.__check_username(name) and self.__check_password(passd):
            return True
        else:
            return False




   
    
```

```python
PM=Password_Manager("varoon","parrot77")
print("Login succesfully?",PM.verify_method("varoon","parrot77"))
print("login sucessfully?",PM.verify_method("varoon","carrot88"))
      
PM.set_password("wrongpass","cheku")
print("login with new password?",PM.verify_method("varoon","cheku"))
```
