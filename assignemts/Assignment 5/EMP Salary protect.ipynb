```python
class Employee:
    def __init__(self,name,salary):
        self.__name=name
        if salary>0:
            self.__salary = salary
        else:
            self.__salary = 0
            print("Invalid salary!")

    def get_name(self):
        return self.__name
    def set_salary(self,salary):
        if salary >0:
            self.__salary = salary
        else:
            print("salary must be in positive need!")

    def show_details(self):
        print("Name:",self.__name)
        print("Salary:",self.__salary)
        
            
        
    
```

```python
e1 = Employee("Varoon",-10000)
e1.show_details()
e1.set_salary(-10000)
```
