```python
import numpy as np
```

```python
#create array  from 1 to 12
a = np.arange(1,13)
print("a (1.....12) :",a)
```

```python
# reshape into 3x4
b=a.reshape(3,4)
print("\n (3,4):\n",b)
```

```python
#reshape it into 2x6
c = b.reshape(2,6)
print("\n (2,6):\n",c)
```

```python
#task 2 flatten() , ravel() , on 2D array c
f_copy = c.flatten()
f_view = c.ravel()

print("\n f_copy ( form c.flatten()):", f_copy)
print ("\n f_copy (from c.ravel()):", f_view)
```

```python
# modify 
f_view[0]= 44
print("\n After f_view[0] = 44")
print("f_view:",f_view)
print("  c now become :\n",c)
```

```python
# Given arr = np.arange(1,10).reshape(3,3)
arr = np.arange(1,10).reshape(3,3)
print("\n arr:\n",arr)
```

```python
arr_f = arr.flatten()
arr_rav= arr.ravel()

print("\n arr_f :",arr_f)
print("\n arr_rav :",arr_rav)
```

```python
# change arr_f 
arr_f[0] = -44
print("\n After arr_f[0] = -44")
print("arr_flat", arr_f)
print(" arr now:\n",arr)
```

```python
# change arr_rav 
arr_rav[0] = -44
print("\n After arr_rav[0] = -44")
print("arr_rav:", arr_rav)
print(" now arr:\n", arr)
```
